```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ranks = [cd_ranks, p_ranks, ml_ranks, mm1_ranks, mm2_ranks, sl_ranks, sm1_ranks, sm2_ranks] = np.load('1cdata')
```

```python
cd_cumsum_pct = np.cumsum(cd_ranks) / np.cumsum(cd_ranks)[-1]
p_cumsum_pct = np.cumsum(p_ranks) /np.cumsum(p_ranks)[-1]
ml_ranks_pct = np.cumsum(ml_ranks) / np.cumsum(ml_ranks)[-1]
mm1_ranks_pct = np.cumsum(mm1_ranks) / np.cumsum(mm1_ranks)[-1]
mm2_ranks_pct = np.cumsum(mm2_ranks) / np.cumsum(mm2_ranks)[-1]
sl_ranks_pct = np.cumsum(sl_ranks) / np.cumsum(sl_ranks)[-1]
sm1_ranks_pct = np.cumsum(sm1_ranks) /np.cumsum(sm1_ranks)[-1]
sm2_ranks_pct = np.cumsum(sm2_ranks) / np.cumsum(sm2_ranks)[-1]
```

```python
fetus=[cd_cumsum_pct[20000], p_cumsum_pct[20000], ml_ranks_pct[20000], mm1_ranks_pct[20000], 
       mm2_ranks_pct[20000], sl_ranks_pct[20000], sm1_ranks_pct[20000], sm2_ranks_pct[20000]]
```

```python
N = 8

fig, ax = plt.subplots(figsize=(5, 4))
ind = np.arange(N)
width = 0.35
p1 = ax.bar(ind, fetus, width, color=['blue', 'orange', 'green', 'green', 'green', 'red', 'red', 'red'])

# ax.set_title("Pearson's $r$ by category")
ax.set_xlabel('Model')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('C:D', 'Parallelogram', 'MAML-Linear', 'MAML-MLP1', 'MAML-MLP2', 'SGD-linear', 
                    'SGD-MLP1', 'SGD-MLP2'), rotation='vertical')
ax.grid(axis='y')

ax.set_ylabel("Cumulative proportion of human \nresponses retrieved in top 1%")
# ax.set_ylim(bottom=-0.4, top=0.6)
ax.autoscale_view()
```

```python
pcts = np.arange(len(cd_cumsum_pct)) / len(cd_cumsum_pct) * 100
```

```python
fig, ax = plt.subplots(figsize=(5,4))
ax.plot(pcts, cd_cumsum_pct, label='C:D similarity')
ax.plot(pcts, p_cumsum_pct, label='Parallelogram')
ax.plot(pcts, mm2_ranks_pct, label='MAML-MLP2')
ax.plot(pcts, sm1_ranks_pct, label='SGD-MLP1')

ax.set_xscale('log')
ax.set_xlim((1e-1, 1e2))
ax.set_xlabel('Percentage (log)')
ax.set_ylabel('Cumulative proportion of human \nresponses retrieved')
ax.legend(loc='lower right')
ax.grid()
```
